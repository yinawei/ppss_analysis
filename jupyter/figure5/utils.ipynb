```python
import os
import pandas as pd
import random
import numpy as np
random.seed(42)

obj_dir = '../../data/hippo/obj/swc'
ext = '.obj'
exist_files = {f[:-len(ext)] for f in os.listdir(obj_dir) if f.endswith(ext)}
print(len(exist_files))
```

```python
res_df = pd.read_csv('../../data/CA1_hippo_data.csv')
res_df
```

```python
res_df = res_df[res_df['Unnamed: 0'].isin(exist_files)]
res_df
```

```python
def sample_per_group(g):
    return g.head(200) if len(g) > 200 else g

selected = (res_df.groupby('label', group_keys=False)
                  .apply(sample_per_group)
                  .reset_index(drop=True))

selected
```

```python
out_df = selected[['Unnamed: 0']].copy()
out_df = out_df.rename(columns={'Unnamed: 0': 'swc_id'})

import random

def rand_color():
    return '#{:06x}'.format(random.randint(0, 0xFFFFFF))

out_df['color'] = [rand_color() for _ in range(len(out_df))]
out_df.to_csv('../../data/selected_obj_200perCluster.csv', index=True)

out_df['folder'] = 'cluster_' + selected['label'].astype(str)

out_df.to_csv('../../data/selected_obj_200perCluster.csv', index=True)
```

```python
import os, shutil, random, pandas as pd

SRC_DIR = '../../data/hippo/obj/swc'
CSV_FILE = '../../data/selected_obj_200perCluster.csv'
N_CLASS = 2
EXT = '.obj'

picked_folders = ['cluster_1', 'cluster_2']
df = pd.read_csv(CSV_FILE)

for i, folder_name in enumerate(picked_folders, 1):
    dst_dir = f'../../data/hippo/obj/cluster_{i}'
    os.makedirs(dst_dir, exist_ok=True)

    for swc_id in df.loc[df['folder'] == folder_name, 'swc_id']:
        src = os.path.join(SRC_DIR, swc_id + EXT)
        dst = os.path.join(dst_dir,   swc_id + EXT)
        if os.path.isfile(src):
            shutil.copy2(src, dst)
        else:
            print(src)
```
