```python
import pickle
from scipy.optimize import curve_fit
def linear_function(x, a, b):
    return a * x + b
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
from scipy.cluster.hierarchy import dendrogram, linkage, leaves_list, ward
from scipy.spatial.distance import pdist
import pickle
import numpy as np
import shutil
from scipy.cluster.hierarchy import dendrogram, linkage, leaves_list, fcluster
from sklearn import metrics
from scipy import stats
```

```python
path_df = pd.read_csv('../../data/ccf_region_k_hop.csv',
                      sep=',', index_col=0)
path_df
```

```python
pt_df = path_df.pivot_table(index='source_region', 
                            columns= ['target_region'], 
                            values=['path_length'], 
                            fill_value=-1,
                            aggfunc='mean'
                           )
pt_df.index.name = ''
pt_df.columns = [i[-1] for i in pt_df.columns]
pt_df
```

```python
v = pt_df.values.reshape(-1)
v = v[v>-1]
sns.histplot(v)
```

```python
region_sim = pd.read_csv('../figure1/eucl_region_similarity_using174_region.csv', sep=',', index_col=0)

region_sim
```

```python
intersection_list = list(set(region_sim.index).intersection(set(path_df['source_region'])))
print(intersection_list[0:5])
```

```python
path_df2 = path_df[path_df['source_region'].isin(intersection_list) & path_df['target_region'].isin(intersection_list)].copy()
region_sim = region_sim.loc[intersection_list, intersection_list].copy()
path_df2['sim'] = -1

for i in path_df2.index:
    tmp_src = path_df2.loc[i, 'source_region']
    tmp_tar = path_df2.loc[i, 'target_region']
    path_df2.loc[i, 'sim'] = region_sim.loc[tmp_src, tmp_tar]

path_df2 = path_df2[path_df2['source_region'] != path_df2['target_region']].copy()

path_df2
```

```python
path_df2['sim_cat'] = -1
for i in list(range(1,21))[::-1]:
    path_df2.loc[path_df2['sim']<i/20, 'sim_cat'] = i

path_df2 = path_df2[path_df2['sim_cat'] != -1]
path_df2
```

```python
stats.pearsonr(path_df2['path_length'], path_df2['sim'])
```

```python
cross_df = pd.crosstab(path_df2['path_length'], path_df2['sim_cat'])
# cross_df.index.name = ''
# cross_df.columns.name = ''

# cross_df = cross_df.iloc[:0:-1, :0:-1].copy()

cross_df
```

```python
x = path_df2['sim_cat']
y = 7-path_df2['path_length']

params, covariance = curve_fit(linear_function, x, y)
a, b = params
x_fit = np.linspace(min(x)-1, max(x), 100)
y_fit = linear_function(x_fit, a, b)
```

```python
x_fit
```

```python
f, ax = plt.subplots(1,1,figsize=(10,3))
sns.heatmap(cross_df.iloc[::-1,:], 
            cmap=sns.color_palette("light:royalblue", as_cmap=True),)
plt.plot(x_fit, y_fit, color='red', 
         linewidth=2, linestyle='--',
         label='Fitted Curve')
plt.savefig('./k_hop_neighbor.svg', bbox_inches='tight')
stats.spearmanr(path_df2['path_length'], path_df2['sim'])
```
