```python
import os
import pandas as pd
import numpy as np
import pickle
from tqdm import tqdm
import seaborn as sns
from scipy.spatial.distance import cdist
from scipy import stats, spatial
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
#plt.rcParams['font.family'] = 'Arial'

from scipy.optimize import curve_fit

def linear_function(x, a, b):
    return a * x + b
```

```python
df = pd.read_csv('../../data/compare_vasculature_data.csv', 
                 sep=',', index_col=0)
df
```

```python
pac_summary = pd.read_csv('../../data/ppss_from_pac_summary.csv', sep=',', index_col=0)
pac_summary
```

```python
meta = pd.read_csv("../../data/Mouse.csv")
meta
```

```python
synapse_num = np.zeros((len(df.index, )))
for i, region in enumerate(df.index):
    flag = False
    for meta_region in meta["Acronym"].values:
        if region == meta_region and flag == False:
            temp = meta[meta["Acronym"] == meta_region].iloc[0, 10:]
            region_level_meta = np.where(~pd.isna(temp))[0][0]

            if meta_region in pac_summary.index.values:
                synapse_num[i] += pac_summary.loc[meta_region].sum()

            flag = True
        elif region != meta_region and flag == False:
            continue
        elif flag:
            temp = meta[meta["Acronym"] == meta_region].iloc[0, 10:]
            region_level = np.where(~pd.isna(temp))[0][0]

            if region_level == region_level_meta:
                flag = False
                break
            else:
                if meta_region in pac_summary.index.values:
                    synapse_num[i] += pac_summary.loc[meta_region].sum()
        else:
            continue
            
synapse_num
```

```python
df['#synapse'] = synapse_num

df
```

```python
df['log_length'] = np.log10(df['local vascular length'])
df['log_bif'] = np.log10(df['number of bifurcation'])
df['log_synapse'] = np.log10(df['#synapse'])

df
```

```python
x = df['log_synapse']
y = df['log_length']

params, covariance = curve_fit(linear_function, x, y)
a, b = params
x_fit = np.linspace(min(x)-1, max(x)+5, 100)
y_fit = linear_function(x_fit, a, b)
```

```python
print(stats.pearsonr(df['log_length'], df['log_synapse']))
sns.scatterplot(data=df,y='log_length', x='log_synapse')
plt.ylim([2.49, 2.92])
plt.xlim([2.5, 7.2])
plt.yticks([2.5, 2.6, 2.7, 2.8, 2.9])
plt.xticks([3, 4, 5, 6, 7])

plt.plot(x_fit, y_fit, color='red', 
         linewidth=2, linestyle='--',
         label='Fitted Curve')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# plt.savefig('./cmp_vasculature.jpg', dpi=400, bbox_inches='tight')
plt.savefig('./cmp_vasculature.svg', bbox_inches='tight')
```
